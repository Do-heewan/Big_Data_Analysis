# 데이터 시각화만을 위한 기법 : t-SNE

```python
from sklearn.manifold import TSNE

tsne = TSNE(n_components=2, perplexity=30) # 초기화
result = tsne.fit_transform(dataset) # 학습
```

### 주요 파라메터

N_Componenets : 기본값 2, 차원 축소 결과 임베딩 자원

Perplexity : 기본값 30, Nearest neighbor의 범위를 설정하는 값

Learning_rate : 기본값 200, 학습할 때 사용되는 lr

N_iter : 기본값 1000, 학습 횟수

N_iter_without_progress : 기본값 300, 조기 종료 옵션(early stop)

````python
from sklearn.manifold import TSNE
from sklearn.datasets import load_digits

import matplotlib.pyplot as plt

digits = load_digits()

fig, axis = plt.subplots(2, 5, figsize = (10, 5), subplot_kw = {'xticks' : (), 'yticks' : ()})

for ax, imgs in zip(axis.ravel(), digits.images):
    ax.imshow(imgs)
````

````python
colors = ['red', 'blue', 'green', 'yellow', 'purple', 'orange', 'pink', 'brown', 'gray', 'black']

tsne = TSNE(perplexity=30)
digits_tsne = tsne.fit_transform(digits.data)
print(digits_tsne.shape)

plt.figure(figsize = (10, 10))
plt.xlim(digits_tsne[:, 0].min(), digits_tsne[:, 0].max() + 1)
plt.ylim(digits_tsne[:, 1].min(), digits_tsne[:, 1].max() + 1)

for i in range(len(digits.data)):
    plt.text(digits_tsne[i, 0], digits_tsne[i, 1], str(digits.target[i]), color = colors[digits.target[i]], fontdict = {'weight' : 'bold', 'size' : 9})

plt.xlabel("TSNE feature 0")
plt.ylabel("TSNE feature 1")
````
